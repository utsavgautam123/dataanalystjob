```python
import numpy as np
df['Founded'] = df['Founded'].replace(-1, np.nan)
df['Founded'] = df['Founded'].fillna(df['Founded'].median())
df['Founded'] = df['Founded'].astype(int)
```

```python
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace(r'\(.*?\)', '', regex=True).str.strip()
df['min_salary'] = df['Salary_Estimate'].str.extract(r'\$(\d+)K')[0].astype(int) * 1000
df['max_salary'] = df['Salary_Estimate'].str.extract(r'-\$(\d+)K')[0].astype(int) * 1000
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2
```

```python

df['avg_salary'] = df['avg_salary'].astype(int)
```

```python
from matplotlib.ticker import FuncFormatter
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(14,6))
k_formatter = FuncFormatter(lambda x, pos: f'${int(x/1000)}K')
ax1.xaxis.set_major_formatter(k_formatter)
sns.histplot(df["avg_salary"],kde=True,ax=ax1,color="red")
mean_val=df["avg_salary"].mean()
median_val=df["avg_salary"].median()
ax1.axvline(mean_val, color='blue', linestyle='--', label=f'Mean: ${mean_val:,.0f}')
ax1.axvline(median_val,color="green",linestyle="-",label=f'median: ${median_val:,.0f}')
ax1.legend()
ax1.set_xlabel("Salary", fontweight='bold')
ax1.set_ylabel("Frequency", fontweight='bold')
ax1.set_title("Salary Distribution with Stats", fontweight='bold')

for label in ax1.get_xticklabels():
    label.set_fontweight('bold')
    
sns.boxplot(y=df["avg_salary"],color="lightgreen",ax=ax2) 
ax2.yaxis.set_major_formatter(k_formatter) 
ax2.set_title("salary Range & Outliers",fontweight="bold")
  
plt.tight_layout()
plt.show()
```

```python
bins=[0, 2, 3, 3.5, 4, 4.5, 5]
labels=['<2', '2-3', '3-3.5', '3.5-4', '4-4.5', '4.5-5']
df["Rating range"]=pd.cut(df["Rating"],bins=bins,labels=labels)
print(df[['Rating', 'Rating range']].head())
ratingcol = df.groupby("Rating range")["avg_salary"].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(12, 6))


bar=sns.pointplot(data=ratingcol, x='Rating range', y='avg_salary',color="lightgreen")
plt.line_label(bar.containers[0], labels=[f'${int(x/1000)}K' for x in ratingcol['avg_salary']], 
             padding=3, fontweight='bold')


k_formatter = FuncFormatter(lambda x, pos: f'${int(x/1000)}K')
plt.gca().yaxis.set_major_formatter(k_formatter)

plt.title('Average Salary by Rating range', fontsize=14, fontweight='bold')
plt.xlabel('Rating range', fontweight='bold')
plt.ylabel('Average salary', fontweight='bold')
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

```
